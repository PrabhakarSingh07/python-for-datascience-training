# Exploratory data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/Housing/Bangalore.csv"
```

```python
df = pd.read_csv(path)
```

Average price of house in bangalore

```python
avg_price = np.mean(df['Price'])
```

```python
print("Average price of house in banagalore: INR. ",avg_price)
```

```python
np.std(df['Price'])
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="Price")
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(data=df,x='Price')
```

```python
df['Location'].value_counts()
```

```python
avg_df = df.groupby('Location',as_index=False).mean()[["Location","Price"]]
```

```python
avg_df = avg_df.sort_values(by="Price",ascending=False)
```

```python
avg_df.head()
```

```python
avg_df.tail()
```

```python
cols = ['Sompura','Krishna Garden Road','Domlur Layout','Richmond Town','Bannerghatta Road Jigani','Kalkere','Basavanagudi']
```

```python
n_df = df.loc[df['Location'].isin(cols)][["Price","Area","Location"]]
```

```python
n_df.head()
```

```python
sns.boxplot(data=n_df,y='Location',x='Price')
```
